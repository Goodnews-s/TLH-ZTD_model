```python
def Get_ZTD(need,path):
    ZTD=[]
    data=pd.read_excel(path)
    for i in range(len(need)):
        ZHD = (0.0022768 * need[i][0])/(1 - 0.0026*np.cos(2*np.deg2rad(data['lat'][i]))-0.00000028*data['hig'][i])
        ZWD = 10**(-6)*(16.52+(k3/need[i][1]))*(287.0464/((need[i][2]+1)*gm))*need[i][3]
        ZTD.append(ZHD+ZWD)
    return ZTD
```

```python
def Get_xlsx(path,outpath):
    need=Get_need(path)
    ZTD=Get_ZTD(need,path)
    data=pd.read_excel(path)
    infor=data.columns.tolist()
    infor.insert(6,'gpt3_ztd')
    data=data.reindex(columns=infor)
    data['gpt3_ztd']=ZTD
    data.to_excel(outpath,index = False)
```
