```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import pandas as pd
import numpy as np
from numpy import *
import pandas as pd
from sklearn.metrics import mean_squared_error, explained_variance_score
import time
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from keras.layers import LSTM, Bidirectional, Dense
from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dense, Conv1D, Flatten, LeakyReLU, Dropout, Input, BatchNormalization
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
```

```python
xlsbpath=r"D:\\Xinshuai\\文献\\老大宝\\six_comp\\part1_comp\\lssvm"
os.chdir(xlsbpath)
filelist= os.listdir(xlsbpath)
scaler = MinMaxScaler()
kernel_size = 2
dropout = 0.2
```

```python
pre=[]
for k in range(len(filelist)):
    df=pd.read_excel("D:\\Xinshuai\\文献\\老大宝\\six_comp\\part1_comp\\lssvm\\"+filelist[k])
    X =df[['lat','lon','hig','lssvm_hgpt_ztd']]
    y = df[['ztd']]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    X_train = np.expand_dims(X_train, axis=2)
    X_test = np.expand_dims(X_test, axis=2)
#     print(X_train)
    
    
    
    model = Sequential()
    model.add(Bidirectional(LSTM(50, return_sequences=True), input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Bidirectional(LSTM(50)))
    

    model.add(Dense(1, activation=LeakyReLU(alpha=0.01)))
    
    
    filepath='D:\\Xinshuai\\文献\\老大宝\\six_comp\\bilstm\\'+filelist[k]+'weights.best.hdf5'
    checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1,save_best_only=True,mode='min',period=1)
    callbacks_list = [checkpoint]
    model.compile(loss='mean_squared_error', optimizer='adam')

    model.fit(x=X_train, y=y_train, batch_size =512,epochs=30,validation_data=(X_test, y_test),verbose=1,callbacks=callbacks_list)
    model=load_model(filepath)
    
    X = scaler.transform(X)
    X = np.expand_dims(X, axis=2)
    pred=model.predict(X)
    pre.append(pred)
    print(filelist[k][:-5]+'站点完成')
```

```python
for i in range(len(filelist)):
    df1=pd.read_excel("D:\\Xinshuai\\文献\\老大宝\\six_comp\\part1_comp\\lssvm\\"+filelist[i])
    df=df1[['time_']]
    df.insert(df.shape[1],'ztd',df1['ztd'])
    df.insert(df.shape[1],'tcn_ztd',df1['tcn_hgpt_ztd'])
    df.insert(df.shape[1], 'bilstm_ztd', pre[i])
    df.to_excel("D:\\Xinshuai\\文献\\老大宝\\six_comp\\bilstm\\"+filelist[i],index=False)
```

```python
tcn_=[]
bilstm_=[]
for k in range(len(filelist)):
    df=pd.read_excel("D:\\Xinshuai\\文献\\老大宝\\six_comp\\bilstm\\"+filelist[k])
    bilstm_.append(np.sqrt(mean_squared_error(df['ztd'], df['bilstm_ztd'])))
    tcn_.append(np.sqrt(mean_squared_error(df['ztd'], df['tcn_ztd'])))
```
